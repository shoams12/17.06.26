```python
import pandas as pd
import numpy as np
import io
```

```python
csv_data = """order_id,customer,city,total_bill,tip,table_size
1,Alice,Tel Aviv,45.50,8.00,4
2,Bob,Haifa,22.30,3.50,2
3,Carol,Tel Aviv,67.80,12.00,6
4,David,Jerusalem,15.20,2.00,2
5,Eve,Haifa,89.90,15.00,8
6,Frank,Tel Aviv,33.40,5.50,3
7,Grace,Jerusalem,51.00,9.00,4
8,Henry,Haifa,28.60,4.00,2
9,Iris,Tel Aviv,74.20,13.50,5
10,Jake,Jerusalem,19.80,2.50,2"""
```

```python
#1
# Load CSV string into DataFrame
df = pd.read_csv(io.StringIO(csv_data))

# Display DataFrame information
print(df.info())

# Display descriptive statistics transposed
print(df.describe().transpose())
```

```python
#2
df['price_per_person'] = df['total_bill'] / df['table_size']
df
```

```python
#3
df[df['price_per_person'] > 12]
```

```python
#4
def size_s_l(table_size):
    return 'Large' if table_size >= 4 else 'Small'

df['table_type'] = df['table_size'].apply(size_s_l)
df
```

```python
#5
df = df.set_index('customer')
df.loc['Alice': 'Carol']
```

```python
#6
l_f_3 = pd.concat(
    [df.head(3), df.tail(3)],
    axis=0,
    ignore_index=True   # resets the index to 0,1,2,3,4
)
l_f_3
```

```python
#7
def value_label(total_bill, tip):
    return 'High Value' if total_bill > 50 and tip > (total_bill * 10) / 100 else 'Standard'

df['value_label'] = np.vectorize(value_label)(
    df['total_bill'],
    df['tip']
)
df
```
